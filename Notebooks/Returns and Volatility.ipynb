```python
import pandas as pd
import numpy as np
from math import sqrt
from google.colab import files
```

```python
# Define the path to the Excel file
excel_file_path = '/content/drive/MyDrive/Finance project Python/Data/ZRR1 Comdty (Rapeseed).xlsx'

# Import the Excel file into a pandas DataFrame
df = pd.read_excel(excel_file_path)

# Calculate Returns
df['Returns'] = np.log(df['ZRR1 Comdty (USD)  (R2)']) - np.log(df['ZRR1 Comdty (USD)  (R2)'].shift(1))

# Calculate Volatility
trading_days_per_year = 252
df['Volatility'] = sqrt(trading_days_per_year) * df['Returns'].std()

# Save the cleaned data to a new Excel file
file_name = 'ZRR1 Comdty Comdty Return_Volatility.xlsx'
df.to_excel(file_name, index=False)

# Download the cleaned file
files.download(file_name)
```

```python
# Read the Excel file into a pandas DataFrame
data = pd.read_excel('/content/drive/MyDrive/Finance Project 2/WeekendCleanedEPU.xlsx')

# Sort the data by date in ascending order
data = data.sort_values('Date')

# Calculate the logarithmic level of EPU
data['Logarithmic Level of EPU'] = np.log(data['EPU'])

# Save the cleaned data to a new Excel file
file_name = 'Log(EPU).xlsx'
data.to_excel(file_name, index=False)

# Download the cleaned file
files.download(file_name)
```
