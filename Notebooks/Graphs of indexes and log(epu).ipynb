```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('/content/drive/MyDrive/Finance Project 2/2 index combinations/B01 Returns and log(EPU).xlsx')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
index1 = df['BO1 Comdty Returns']
index2 = df['Logarithmic Level of EPU']
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], index1, label='BO1 Comdty Returns')
plt.plot(df['Date'], index2, label='Logarithmic Level of EPU')
plt.xlabel('Date')
plt.ylabel('Index Value')
plt.title('Comparison of Index 1 and Index 2')
plt.legend()
plt.grid(True)
plt.show()
```
